### Grundlegendes zur Sonne

Die Sonne ist etwa 4,6 Milliarden Jahre alt und vereint 99,86% der Masse unseres Sonnensystems. Sie hat einen Durchmesser von 1,39 Mio. km, was etwa dem 109-fachen Durchmesser der Erde entspricht. Die Sonne besteht zu 73% aus Wasserstoff und zu 25% aus Helium. Die restlichen 2% sind Metalle. Im Kern der Sonne herrschen Drücke von 200 Milliarden bar und Temperaturen bis zu 15,6 Mio. K.

Folgende Bilder zeigen die aktuelle Ansticht der Sonne in zwei engen Wellenlängenbereichen im Extremen Ultraviolett (EUV). Sie stammen vom [NASA-Satelliten SDO](https://sdo.gsfc.nasa.gov), der im Sekundentakt Bilder der Sonne aufnimmt.

 $\lambda=$ 17,1 nm | $\lambda=$ 30,4 nm
:-------:|:--------:
 <img align="left" src="https://sdo.gsfc.nasa.gov/assets/img/latest/latest_512_0171.jpg" width="300"> | <img align="left" src="https://sdo.gsfc.nasa.gov/assets/img/latest/latest_512_0304.jpg" width="300"> 
 
Im Inneren der Sonne findet eine Fusionsreaktion statt, bei der jeweils vier Wasserstoffkerne (Protonen) zu einem Heliumkern verschmelzen. Der entstehende Heliumkern hat eine um 0,75% geringere Masse als die vier Wasserstoffteilchen, so dass die Masse der Sonne insgesamt um $\dot{m} =$ 4,3 Mio. t pro Sekunde abnimmt.

Nach der von Einstein entdeckten Äquivalenz von Energie und Masse

$$ E = m c^2$$

bzw. von Leistung und Massenstrom

$$ \Phi = \dot{m} c^2$$

kann dieser [Massendefekt](https://de.wikipedia.org/wiki/Massendefekt#Massendefekt_des_Atomkerns_von_4He) in eine Leistung umgerechnet werden, die von der Sonne in Form von Strahlung an das Weltall abgegeben wird:

<div class=highlight>
$$ \Phi_\text{Sonne}  = 4,28\cdot 10^9 \ \frac{\text{kg}}{\text{s}}\cdot \left(299,8\cdot 10^6 \ \frac{\text{m}}{\text{s}}\right)^2 = 3,845\cdot 10^{26}\ \text{W}$$
</div>

Die pro Sekunde von der Sonne abgestrahlte Energie würde ausreichen, um den gesamten Primärenergiebedarf der Menschheit für 750 000 Jahre zu decken.

#### Welcher Anteil der Strahlungsleistung trifft auf die Erdoberfläche?

Da die Sonne in alle Raumrichtungen Strahlung emittiert kommt davon natürlich nur ein Teil auf der Erde an. Denkt man sich eine Kugel deren Radius dem mittleren Abstand zwischen Sonne und Erde entspricht (149,6 Mio. km), so muss die gesamte Strahlungsleistung durch deren Oberfläche treten. Teilt man die gesamte Leistung durch die Oberfläche der Kugel, so erhält man die spezifische Strahlungsleistung, die pro m$^2$ auf die Erde trifft:

<div class=highlight>
$$E_0 = \frac{\Phi_\text{Sonne} }{4\pi r_\text{SE}^2} = \frac{3,845\cdot 10^{26}\ \text{W}}{4\pi \left(149,6\cdot 10^{9}\right)^2 \text{m}^2} = 1367 \frac{\text{W}}{\text{m}^2}$$
</div>

Der Wert 1367 W/m$^2$ ist auch als [Solarkonstante](https://de.wikipedia.org/wiki/Solarkonstante) bekannt. Er lässt sich außerhalb der Erdatmosphäre mit einem Messgerät, das senkrecht zur Sonneneinstrahlung angeordnet ist sehr genau messen. Da sich der Abstand zwischen Sonne und Erde innerhalb eines Jahres ändert, variiert auch der Wert für die Bestrahlungsstärke zwischen 1325 W/m$^2$ und 1420 W/m$^2$.

<img src="sonne-erde.png" width="500">

Multipliziert man die Solarkonstante mit der projizierten Kreisfläche der Erde, so erhält man die Gesamtstrahlungsleistung, die auf die Erde trifft. Mit dem Durchmesser der Erde $d_\text{Erde}$ = 12.742 km ergibt das

<div class=highlight>
$$\Phi_{\text{Erde}} = E_0 \cdot \frac{\pi}{4} d_\text{Erde}^2 = 1,743\cdot 10^{17} \ \text{W} = 174,3 \ \text{PW}$$
</div>

Der jährliche Primärenergieverbrauch weltweit liegt derzeit (Stand 2016) bei rund 160 PWh. D.h. die pro Stunde auf die Erde einstrahlende Sonnenenergie wäre rein rechnerisch ausreichend, um die Menschheit ein Jahr mit Energie zu versorgen. Da ein großer Teil der Sonnenenergie zum Antrieb des Wasserkreislaufs und der Windsysteme auf der Erde dient, hinkt dieser Vergleich natürlich etwas.

#### Spektrum der Sonnenstrahlung

Unter der Annahme, dass sich die Sonne wie ein [Schwarzkörperstrahler](https://de.wikipedia.org/wiki/Schwarzer_Körper) verhält, lässt sich die Temperatur der Sonne mit dem [Stefan-Bolzmann-Gesetz](https://de.wikipedia.org/wiki/Stefan-Boltzmann-Gesetz) berechnen

$$\Phi_\text{Sonne} = \sigma \cdot A_\text{Sonne} \cdot T_\text{Sonne}^4$$

mit der Stefan-Bolzmann-Konstanten $\sigma = 5,67037\cdot 10^{-8}\  \frac{\text{W} }{\text{m}^2\text{K}^4}$ und dem Sonnendurchmesser von 1,39 Mio. km ergibt sich aufgelöst nach der Temperatur

<div class=highlight>
$$T_\text{Sonne} = \left(\frac{\Phi_\text{Sonne}}{\sigma\cdot\pi\cdot d_\text{Sonne}^2 }\right)^\frac{1}{4} = 5778\ \text{K}$$
</div>

Mit der nun bekannten Temperatur, kann das Spektrum der Sonnenstrahlung mit dem [Planck'schen Strahlungsgesetz](https://de.wikipedia.org/wiki/Plancksches_Strahlungsgesetz) berechnet werden. Die spektrale Bestrahlungsstärke

<div class=highlight>
$$M(\lambda,T)_\text{Sonne}  = \frac{2\pi h c^2}{\lambda^5}\cdot \frac{1}{\text{exp}\left(\frac{h c}{\lambda k T}\right) - 1 }$$
</div>

hat die Einheit Leistung pro Fläche und Wellenlänge (W/(m$^2 \mu$m)).

Um damit die auf die Erdatmosphäre treffende spektrale Bestrahlungsstärke zu berchnen, muss die Gleichung noch mit der Sonnenoberfläche multipliziert und durch die Kugeloberfläche mit Radius Sonne-Erde geteilt werden.

<div class=highlight>
$$M(\lambda,T)_\text{Erde} = M(\lambda,T)_\text{Sonne} \frac{d_\text{Sonne}^2}{4 r_\text{SE}^2 }$$
</div>

Das folgende Diagramm zeigt das [so berechnete](sonne.ipynb) Spektrum als schwarze Linie. Zum Vergleich ist das von einem Satelliten außerhalb der Erdatmosphäre gemessene Spektrum in rot dargestellt. Dieses Spektrum wird auch AM0-Spektrum genannt. Die Bezeichnung steht für "air mass zero", d.h. es ist keine absorbierende Atmosphäre im Strahlengang vorhanden.

```python
%run -i "./sonne.ipynb"
plotAM0()
```

Wie aus dem Diagramm hervorgeht, liegt das Intensitätsmaximum bei einer Wellenlänge von etwa 500 nm, was grünem, sichtbarem Licht entspricht. Das menschliche Auge kann Strahlung im Wellenlängenbereich von etwa 400 nm (violett) bis 700 nm (rot) wahrnehmen und weist bei 500 nm die größte Empfindlichkeit auf.

### Einfluss der Atmosphäre

Die Bestandteile der Atmosphäre, insbesondere Wasserdampf (H$_2$O), Ozon (O$_3$), Sauerstoff (O$_2$) und Kohlendioxid (CO$_2$), absorbieren einen Teil der Sonnenstrahlung. Außerdem werden die blauen Frequenzanteile an den Molekülen gestreut und abgeschwächt. Insgesamt ergibt sich dadurch auf der Erdoberfläche ein von AM0 abweichendes Spektrum.

Wie stark die jeweiligen Wellenlängenbereiche absorbiert werden hängt davon ab, welche Strecke das Sonnenlicht durch die Atmosphäre zurücklegt. Das folgende Diagramm zeigt das Wellenlängenspektrum auf der Erdoberfläche bei einem Einfallswinkel von $\gamma_S = 41,8^\circ$. Im Vergleich zur senkrechten Einstrahlung legt dass Sonnenlicht bei diesem Winkel die 1,5-fache Strecke zurück. Man spricht deshalb auch vom **AM1.5-Spektrum**.

```python
plotAM15()
```

Aus dem Diagramm wird deutlich, dass die harte UV-Strahlung mit Wellenlängen kleiner 300 nm nicht mehr im Spektrum vertreten sind. Ursächlich hierfür ist die **Ozonschicht**, in der die Strahlung mit kleinen Wellenlängen fast vollständig absorbiert wird.

Die **Treibhausgase** CO$_2$, H$_2$O, N$_2$O, CH$_4$ sowie FCKW absorbieren bestimmte Strahlungsbänder im Infrarotbereich ($\lambda > 750\ $nm) und führen zu Einbrüchen im Spektrum.

Durch [**Rayleigh-Streuung**](https://de.wikipedia.org/wiki/Rayleigh-Streuung) an Luftmolekülen, die einen kleineren Durchmesser als die Wellenlänge aufweisen, wird das einfallende Licht gestreut und abgeschwächt. Da der [Streuquerschnitt](https://de.wikipedia.org/wiki/Wirkungsquerschnitt) proportional zu $\lambda^{-4}$ ist, wird blaues Licht stärker gestreut als rotes. Die geringere Intensität bei kleinen Wellenlängen ist in erster Linie auf diesen Effekt zurückzuführen. 

Die Rayleigh-Streuung ist auch der Grund dafür, dass der Himmel blau und die Sonne, wenn sie senkrecht am Himmel steht nicht weiß sondern gelb erscheint. Die blauen Strahlungsanteile werden stärker in alle Himmelsrichtungen gestreut (-> blauer Himmel) und es bleiben mehr rötliche Anteile (-> gelbe Sonne). Am Morgen und am Abend, ist der Einfallswinkel kleiner und die Strahlung legt einen größeren Weg in der Atmosphäre zurück. Es werden dann noch mehr blaue Anteile gestreut und die Sonne erscheint zunehmend rötlicher.

Durch [**Mie-Streuung**](https://de.wikipedia.org/wiki/Mie-Streuung) an Staubteilchen oder Aerosolen, deren Größe etwa der Wellenlänge des Lichts entsprechen wird ebenfalls die Intensität einiger Wellenlängenbereiche reduziert. Der Effekt ist insbesondere in stark industrialisierten Gebieten zu beobachten.

Ein damit verbundener Begriff ist das [Global Dimming](https://de.wikipedia.org/wiki/Globale_Verdunkelung), unterm dem die globale Auswirkung der Luftverschmutzung auf die Sonneneinstrahlung beschrieben wird. Einige Wissenschaftler gehen davon aus, dass der Treibhauseffekt Anfang bis Mitte des letzten Jahrhunderts durch die hohe Luftverschmutzung infolge der Industrialisierung deutlich abgemildert wurde. Durch die Reduktion der Partikelemissionen in den modernen Industriestaaten ist der *global dimming* Effekt heute weniger ausgeprägt, so dass der durch den Menschen verursachte Treibhauseffekt größere Auswirkungen hat.

### Sonnenstand und Sonnenbahn

Da die Rotationsachse der Erde etwa 23,5° gegenüber der Bahnebene um die Sonne geneigt ist, ändert sich der Einfallswinkel des Sonnenlichtes nicht nur über den Tag hinweg zwischen Sonnenauf- und untergang, sondern auch mit der Jahreszeit.

Die folgende Skizze zeigt die geometrische Abhängigkeit vom Standort auf der Erdoberfläche und von der Jahreszeit. Für Orte auf der Nordhalbkugel ist der Einfallswinkel im Juni am höchsten, d.h. das Sonnenlicht legt die geringste Strecke in der Atmosphäre zurück und strahlt mit der größten Intensität auf die Erdoberfläche. Während auf der Nordhalbkugel deshalb Sommer ist, verhält es sich auf der Südhalbkugel gerade anders herum: dort sind die Temperaturen im Juni am niedrigsten.

Orte die näher am Äquator liegen haben steilere Einfallswinkel (hier im Beispiel Gran Canaria).

<img src="erdumlaufbahn.png" width="500">

Um die Position der Sonne von einem Standort auf der Erde genau zu beschreiben werden die Elevations- und Azimutwinkel $\gamma_s$ und $\alpha_s$ verwendet. Ein Elevationswinkel von 90° bedeutet eine senkrechte Einstrahlung (AM1). Der Azimutwinkel beträgt 0° wenn die Sonne genau im Norden steht, 90° für eine Position im Osten, 180° für eine südliche und 270° für eine westliche Position.

<img src="ElevationAzimut.png" width="450">

Mithilfe der beiden Winkel können Sonnenbahndiagramme gezeichnet werden, aus denen für jede Tages- und Jahreszeit die Sonnenposition relativ zur Position auf der Erdoberfläche abgelesen werden kann.
